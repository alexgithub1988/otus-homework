# I. Numpy

### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
first_array = np.zeros(10)
first_array[4] = 1
print(first_array)
t_first_array = first_array.reshape(2,5)
print(t_first_array)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
second_array = np.arange(10,50,1)
print(second_array)
r_second_array = second_array[::-1]
print(r_second_array)
elements = second_array[second_array%2 == 0]
print(elements)
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
d2_array = np.arange(9).reshape(3,3)
print(d2_array)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
third_array =  np.random.random((4,3,2))
print(third_array)
print(third_array.min())
print(third_array.max())
```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
first_arr = np.random.randint(100,size=(6,4))
second_arr = np.random.randint(50,size = (4,3))
print(first_arr)
print(second_arr)
multiply = np.dot(first_arr,second_arr)
print(multiply)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
arr = np.random.randint(100,size = (7,7))
print(arr)
mean = arr.mean()
print(mean)
std_dev = arr.std()
print(std_dev)
norm_arr = (arr - mean)/std_dev
print(norm_arr)
```

### Импортируйте: pandas, matplotlib, seaborn

### Загрузите датасет Tips из набора датасетов seaborn

### Посмотрите на распределение числовых признаков

```python
plt.figure(figsize=(10,5))
plt.title('total_bill')
sns.histplot(tips['total_bill'],kde = True,color = 'green')
plt.show()
```

```python

g=sns.relplot(data=tips,x = 'total_bill', y = 'tip' ,kind = 'scatter' )
g.fig.suptitle('распределение счета и чаевых')
```

### Найдите максимальное значение 'total_bill'

```python
tips['total_bill'].max()
```

### Найдите количество курящих людей

```python
tips[tips['smoker'] == 'Yes']['smoker'].count()
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python
tips.groupby('day',observed = False)['total_bill'].mean()
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python
median_total_bill = tips['total_bill'].median()
filtered_tips = tips[tips['total_bill']>median_total_bill]
avg_from_sex = filtered_tips.groupby('sex',observed = False)['total_bill'].mean()
avg_from_sex

```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
tips['smoker'] = tips['smoker'].map({'Yes':1,'No':0})
tips
```

# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
plt.figure(figsize = (5,5))
sns.histplot(tips['total_bill'])
plt.ylabel('частота')
plt.show()
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize = (10,5))
sns.histplot(data=tips,x='time',y='tip',hue='time')
plt.show()
```
